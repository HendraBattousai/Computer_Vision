```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from PIL import Image, ImageFilter
%matplotlib inline
```

```python
image = cv2.imread('Rafi.jpg',0)
image = cv2.resize(image,(800,800))
```

```python
figure_size = 9
new_image = cv2.medianBlur(image, figure_size)
plt.figure(figsize=(11,6))
plt.subplot(121), plt.imshow(image, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(new_image, cmap='gray'),plt.title('Mean filter')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
kernel_Roberts_x = np.array([
     [1,0],
     [0,-1]                        
])
kernel_Roberts_y = np.array([
     [0,-1],
     [1,0]                        
])

kernel_Sobel_x = np.array([
     [-1, 0, 1],
     [-2, 0, 2],
     [-1, 0, 1]])
kernel_Sobel_y = np.array([
     [1, 2, 1],
     [0, 0, 0 ],
     [-1, -2, -1]])

kernel_Prewitt_x = np.array([
     [-1, 0, 1],
     [-1, 0, 1],
     [-1, 0, 1]])
kernel_Prewitt_y = np.array([
     [1, 1, 1],
     [0, 0, 0],
     [-1, -1, -1]])
```

```python
def Canny(image,k,t1,t2):
  img = cv2.GaussianBlur(image, (k,k),0)
  canny = cv2.Canny(img, t1, t2)
  return canny
```

```python
kernel_Laplacian_1 = np.array([
     [0, 1, 0],
     [1, -4, 1],
     [0, 1, 0]])
kernel_Laplacian_2 = np.array([
     [1, 1, 1],
     [1, -8, 1 ],
     [1, 1, -1]])
kernel_Laplacian_3 = np.array([
     [2, -1, 2],
     [-1, -4, -1],
     [2, 1, 2]])
kernel_Laplacian_4 = np.array([
     [-1, 2, -1],
     [2, -4, 2 ],
     [-1, 2, -1]])

kernel_LoG_4 = np.array([
     [0, 0, -1, 0, 0],
     [0, -1, -2, -1, 0],
     [-1, -2, -16, -2, -1],
     [0, -1, -2, -1, 0],
      [0, 0, -1, 0, 0]])
```
