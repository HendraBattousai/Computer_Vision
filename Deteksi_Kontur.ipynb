# EKSTRAKSI KONTUR

```python
#Contoh untuk mendapatkan tepi objek menggunakan findContours()

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

#Baca citra biner huruf_bin.png
citra = cv2.imread("/content/huruf_bin.jpeg", 0)

#Bentuk citra berwarna dengan Latar belakang hitam
jumBaris = citra.shape[0]
jumKolom = citra.shape[1]
citraKontur = np.zeros((jumBaris, jumKolom, 3), np.uint8)

```

```python
#Memperoleh kontur
kontur,hierarki = cv2.findContours(citra,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)


#Tambahkan kontur
cv2.drawContours(citraKontur, kontur, -1,
                (230, 216, 173), 2)

#Tampilkan hasilnya
citraRGB = cv2.merge((citra, citra, citra))
hasil = np.hstack((citraRGB, citraKontur))
cv2_imshow(hasil)
cv2.waitKey()
```

```python
#Pengaturan warna kontur
import cv2
import numpy as np

#Baca citra biner huruf_bin.png
citra2 = cv2.imread("/content/huruf_bin.jpeg", 0)

#Bentuk citra berwarna dengan Latar belakang hitam
jumBaris = citra.shape[0]
jumKolom = citra.shape[1]
citraKontur = np.zeros((jumBaris, jumKolom, 3), np.uint8)

```

```python
#Memperoleh kontur
kontur, hierarki = cv2.findContours(citra2,
    cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

#Tambahkan kontur dengan warna berlainan
warna = [(230, 216, 173), (255,255,255),
         (255, 255, 0), (255, 0, 255),
         (0, 255, 255)]

for indeks in range(len(kontur)):
    cv2.drawContours(citraKontur, kontur, indeks,
                     warna[indeks % 5], 2)
    
#Tampilkan hasilnya
citraRGB = cv2.merge((citra2, citra2, citra2))
hasil = np.hstack((citraRGB, citraKontur))
cv2_imshow(hasil)
cv2.waitKey()
```

# PENGHAMPIRAN KURVA POLIGON

```python
#Contoh penggunaan approxPolyDB()

import cv2

#Baca citra biner
cupang = cv2.imread("/content/guppy.jpeg", 0)

citraBerwarna = cv2.merge((cupang, cupang, cupang))

#Peroleh kontur
kontur, hierarki = cv2.findContours(cupang,
    cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

print ("Jumlah piksel semula:", len(kontur[0]))

epsilon = 0.01 * cv2.arcLength(kontur[0], True)
approx = cv2.approxPolyDP(kontur[0],epsilon, True)

print ("Jumlah piksel sekarang: ", len(approx))

cv2.drawContours(citraBerwarna, approx, -1,(255, 0, 255), 8)
cv2_imshow(citraBerwarna)
cv2.waitKey()
```

# KOTAK PEMBATAS

```python
# Kotak pembatas

import cv2

#Baca citra biner
cupangkotak = cv2.imread("/content/guppy.jpeg", 0)

citraBerwarna = cv2.merge((cupangkotak, cupangkotak, cupangkotak))

#Peroleh kontur
kontur, hierarki = cv2.findContours(cupangkotak,
    cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

# Peroleh kotak pembatas langsung
x, y, lebar, tinggi = cv2.boundingRect(kontur[0])

# Gambar kotak pembatas
cv2.rectangle(citraBerwarna, (x,y),
              (x + lebar, y + tinggi),
              (255, 255, 0), 2)

# Peroleh kotak pembatas putar
persegiPanjang = cv2.minAreaRect(kontur[0])
kotak = cv2.boxPoints(persegiPanjang)
kotak = np.int32(kotak)
cv2.drawContours(citraBerwarna,[kotak], 0,(128, 128, 128), 2)

# Tampilkan hasilnya
cv2_imshow(citraBerwarna)
cv2.waitKey()
```

# LINGKARAN PEMBATAS

```python
cupangbunder = cv2.imread("/content/guppy.jpeg", 0)
citraBerwarna = cv2.merge((cupangbunder, cupangbunder, cupangbunder))

#Peroleh kontur
kontur, hierarki = cv2.findContours(cupangbunder,
    cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

#Lingkaran Pembatas
(x,y),radius = cv2.minEnclosingCircle(kontur[0])

pusat = (int(x), int(y))
radius = int(radius)

cv2.circle(citraBerwarna,pusat,radius,(255,255,0),2)

#show the result
cv2_imshow(citraBerwarna)
cv2.waitKey()
```

# ELIPS

```python
cupangelips = cv2.imread("/content/guppy.jpeg", 0)
citraBerwarna = cv2.merge((cupangelips, cupangelips, cupangelips))

#Peroleh kontur
kontur, hierarki = cv2.findContours(cupangelips,
    cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

elips = cv2.fitEllipse(kontur[0])
cv2.ellipse(citraBerwarna,elips,(255,255,0),2)

#show the result
cv2_imshow(citraBerwarna)
cv2.waitKey()
```

# COVEX HULL

```python
cupangconvex = cv2.imread("/content/guppy.jpeg", 0)
citraBerwarna = cv2.merge((cupangconvex, cupangconvex, cupangconvex))

#Peroleh kontur
kontur, hierarki = cv2.findContours(cupangconvex,
    cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

convex = cv2.convexHull(kontur[0])
cv2.drawContours(citraBerwarna,[convex],-1,(255,255,0),2)

#show the result
cv2_imshow(citraBerwarna)
cv2.waitKey()
```

##HOUGH CIRCLE TRANSFORMATION

```python
import matplotlib.pyplot as plt

matajiwa = cv2.imread("/content/iris_mata.jpeg", 0 )
c_img = cv2.cvtColor(matajiwa, cv2.COLOR_GRAY2BGR)
canny = cv2.Canny(c_img, 50,50)

circles = cv2.HoughCircles(canny,cv2.HOUGH_GRADIENT,1,1000, param1= 10, param2= 15, minRadius = 20, maxRadius = 40)
circles = np.uint16(np.around(circles))

for i in circles[0,:]:
  #draw the outer circle
  cv2.circle(c_img,(i[0],i[1]), i[1],(0,255,0),2)
  #draw the center of the circle
  cv2.circle(c_img,(i[0],i[2]),2,(10,10,255),3)

plt.imshow(c_img,cmap = 'gray')
plt.show
```
